```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
Combined_Data = pd.DataFrame(school_data_complete)
Combined_Data.head()
```

```python
## Math Reduced DB to help identify Math % per school ##

Reduced_MathDB = Combined_Data.loc[Combined_Data['math_score']>=70]
Math_DB = Reduced_MathDB.groupby('school_name').agg({
    'math_score':'count'
})
Math_DB
```

```python
## Reading DB to help identify Math % per school
Reduced_ReadingDB = Combined_Data.loc[Combined_Data['reading_score']>=70]
Reading_DB = Reduced_ReadingDB.groupby('school_name').agg({
    'reading_score':'count'
    })
Reading_DB
```

```python
Size_DB = Combined_Data.groupby('school_name').agg({
    'size':'mean'
})
Size_DB
```

```python
#Combine Math and Reading Scores

Stem_DB = pd.merge(pd.merge(Math_DB, Reading_DB, on='school_name'), Size_DB, on='school_name')
Stem_DB
```

```python
#Identify Percentages by school
ReadingPerc = Stem_DB['reading_score'] / Stem_DB['size']
MathPerc = Stem_DB['math_score'] / Stem_DB['size']
Stem_DB['Math Passing'] = MathPerc
Stem_DB['Reading Passing'] = ReadingPerc
Stem_DB['Overall Passing'] = (Stem_DB['Math Passing'] + Stem_DB['Reading Passing']) / 2
Stem_DB
```

```python
School_DF = Combined_Data[['school_name','size', 'budget', 'math_score','reading_score',]]
#School_DF = School_DF.drop_duplicates(['type'])
School_DF.head()
```

```python
Grouped_DF = School_DF.groupby('school_name').agg({
    'size':'mean',
    'budget':'mean',
    'math_score':'mean',
    'reading_score':'mean'
})
Grouped_DF
```

```python
AverageCost = Grouped_DF['budget']/Grouped_DF['size']
Grouped_DF["Cost_Per"] = AverageCost
Grouped_DF
```
